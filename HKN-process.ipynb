## **1 판매상품 정보 수집기**

```python
file_item_name = 'data/hknuri_items.xlsx'
file_store_name = 'data/hknuri_store.xlsx'

import pandas as pd
table_items = pd.read_excel(file_item_name)
table_store = pd.read_excel(file_store_name)
table_store.head(3)
```

```python
store_dict = { 
    table_store.loc[_,'상호']:table_store.loc[_,'주소']  
    for _ in range(table_store.shape[0])
}
store_area = { 
    table_store.loc[_,'상호']:table_store.loc[_,'자치구']  
    for _ in range(table_store.shape[0])
}
store_phone = { 
    table_store.loc[_,'상호']:table_store.loc[_,'전화']  
    for _ in range(table_store.shape[0])
}
store_ceo = { 
    table_store.loc[_,'상호']:table_store.loc[_,'대표자']  
    for _ in range(table_store.shape[0])
}

address = [store_dict[_]  for _ in table_items['업체']]
area = [store_area[_]  for _ in table_items['업체']]
phone = [store_phone[_]  for _ in table_items['업체']]
ceo = [store_ceo[_]  for _ in table_items['업체']]
table_items.insert(2, '주소', address)
table_items.insert(2, '자치구', area)
table_items.insert(2, '전화', phone)
table_items.insert(2, '대표자', ceo)
table_items = table_items.sort_values(['자치구','업체','상품']).reset_index(drop=True)
table_items.to_excel('data/hknuri_items_v2.xlsx')
table_items.head(3)
```
