## **2 겹치는 지역들 확인 및 구분하기**

```python
from collections import Counter
lat_multiple = [k for k,v in Counter(table['lat']).items() if v != 1]
_table = table[table['lat'].isin(lat_multiple)]
for lat in lat_multiple:
    _table_temp = _table[_table['lat']==lat]
    count = 0
    for _ in _table_temp.index:
        _table_lon = _table_temp.loc[_, 'lon']
        table.loc[_, 'lon'] = _table_lon - 0.00008 * count
        _table_lat = _table_temp.loc[_, 'lat']
        table.loc[_, 'lat'] = _table_lat - 0.00008 * count
        count += 1

table.to_csv('data/stores_clust_fix.csv', index=None)
map_instance = maplot.plot_circle(table, title='cluster', zoom=15)
map_instance.save('data/map.html')
map_instance
```
